```python
import re

def mood_analysis(text):
  mood_dict = {
      "negatif" : ["yorgun","üzgün","kötü","mutsuz","berbat","iğrenç"],
      "pozitif" : ["mutlu","iyi","enerjik","süper","mükemmel"]
  }
  if any(word in text for word in mood_dict["negatif"]):
    print("🧠 Ruh halin düşük görünüyor. Bugün kendine iyi davran 💙")
  elif any(word in text for word in mood_dict["pozitif"]):
    print("😊 Harika görünüyorsun! Pozitif ruh halin bulaşıcı 💫")

def health_analysis(text):
  health_dict = {
      "hafif": ["başım ağrıyor", "burnum akıyor", "öksürüyorum"],
      "ciddi": ["karın ağrısı", "ateşim var","midem bulanıyor"]
  }
  if any(symptom in text for symptom in health_dict["hafif"]):
        print("🤒 Hafif soğuk algınlığı olabilir. Bol su iç su her derde devadır, C vitamini al. 2 güne geçmezse doktora görün.")
  elif any(symptom in text for symptom in health_dict["ciddi"]):
        print("🔥 Bu durum ciddi olabilir. Bugün dinlenmeni çalışıyorsan işten izin almanı öneririm. Eğer geçmezse doktora gitmelisin.")

def score_analysis(text):
  puanlar=re.findall(r'\d+' , text)
  for puan in puanlar :
    puan =int(puan)
    if puan >= 80:
        print(f"🎉 Harika! {puan} çok iyi bir puan. Gurur duymalısın!")
    elif puan >= 60:
        print(f"👍 Fena değil! {puan} puanla geçtin ama daha iyisini yapabilirsin.")
    else:
        print(f"📚 {puan} biraz düşük. Ama endişe etme, gelişmek için buradayız!")

def main():
  print("Merhaba Kullanıcı Ben LifeBot Ahmet ne derse onu yaparım ")
  print("Ve senden istediğim aşşağıdaki soruların cevabını bana tek seferde yazarmısın")
  print("👉 Bu gün Nasıl hissediyorsun?")
  print("👉 Herhangi bir Rahatsızlık belirtin var mı?")
  print("👉 Yakın zamanda aldığın sınav/performans puanı nedir?")

  girdi = input("\nYanıtınızı Yazın Lütfen :").lower()

  mood_analysis(girdi)
  health_analysis(girdi)
  score_analysis(girdi)

if __name__ == "__main__":
  main()



```
